```python
url =  'https://api.github.com/repos/Ironhack-Data-Madrid-Octubre-2023/apuntes_clase/commits?since=2023-10-28'
```

```python
res = req.get(url).json()
```

```python
res[0]['commit']['author']['date']
```

```python
commits = []



for c in res:
    
    if c['commit']['author']['date'] > '2023-10-28':
        
        commits.append(c)
        
    else:
        continue
```
